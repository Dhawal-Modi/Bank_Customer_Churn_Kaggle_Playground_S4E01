```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
file_path = 'dataset/train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
sns.set(style="whitegrid")

# Creating a figure for multiple subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Histogram of Age
sns.histplot(data=data, x="Age", kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Age")

# Histogram of Credit Score
sns.histplot(data=data, x="CreditScore", kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Distribution of Credit Score")

# Count plot of Geography
sns.countplot(data=data, x="Geography", ax=axes[1, 0])
axes[1, 0].set_title("Count of Customers by Geography")

# Count plot of Gender
sns.countplot(data=data, x="Gender", ax=axes[1, 1])
axes[1, 1].set_title("Count of Customers by Gender")

# Count plot of HasCrCard
sns.countplot(data=data, x="HasCrCard", ax=axes[2, 0])
axes[2, 0].set_title("Count of Customers Having Credit Card")

# Count plot of IsActiveMember
sns.countplot(data=data, x="IsActiveMember", ax=axes[2, 1])
axes[2, 1].set_title("Count of Active Members")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Creating a figure for multiple subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Age vs Exited
sns.boxplot(x="Exited", y="Age", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Age vs Exited")

# CreditScore vs Exited
sns.boxplot(x="Exited", y="CreditScore", data=data, ax=axes[0, 1])
axes[0, 1].set_title("Credit Score vs Exited")

# Geography vs Exited
sns.countplot(x="Geography", hue="Exited", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Geography vs Exited")

# Gender vs Exited
sns.countplot(x="Gender", hue="Exited", data=data, ax=axes[1, 1])
axes[1, 1].set_title("Gender vs Exited")

# Adjust layout
plt.tight_layout()
plt.show()
```
